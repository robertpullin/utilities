# Plotting with `cartopy`

The purpose of this notebook is to explore plotting with `cartopy`.

## Generating Points

```python
from point_generation.random_coordinates import *

coords = generate_random_coordinates(n=10)

lats,lons = coords_to_lat_lon_lists(coords)
```

## Plotting

### Using `matplotlib`

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.scatter(lons, lats);
```

### Using `cartopy` projection

```python
import cartopy.crs as crs

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1,
                     projection=crs.PlateCarree())
ax.scatter(lons, lats);
```

### With coastlines using `ax.coastlines()` and setting size using `figsize`

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1, 1, 1,
                     projection=crs.PlateCarree())
ax.scatter(lons, lats)
ax.coastlines();
```

### Panned using `central_longitude`

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1, 1, 1,
                     projection=crs.PlateCarree(central_longitude=-170))
ax.scatter(lons, lats)
ax.coastlines();
```

### With gridlines using `ax.gridlines()`

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1, 1, 1,
                     projection=crs.PlateCarree(central_longitude=-170))
ax.scatter(lons, lats)
ax.coastlines()
ax.gridlines();
```

### Auto zoom to small area

```python
coords = generate_random_coordinates(n=100,min_lat=20,max_lat=60,min_lon=-90,max_lon=-60)

lats,lons = coords_to_lat_lon_lists(coords)

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1, 1, 1,
                     projection=crs.PlateCarree())
ax.scatter(lons, lats)
ax.coastlines()
ax.gridlines();
```

### Force global view with `ax.set_global()`

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1, 1, 1,
                     projection=crs.PlateCarree())
ax.scatter(lons, lats)
ax.coastlines()
ax.gridlines()
ax.set_global();
```

### Connecting Points

```python
coords = generate_random_coordinates(n=10,min_lat=0,max_lat=60,min_lon=-120,max_lon=-60)

lats,lons = coords_to_lat_lon_lists(coords)

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1, 1, 1,
                     projection=crs.PlateCarree())
ax.plot(lons, lats)
ax.coastlines()
ax.gridlines()
ax.set_global();
```
